### Conditional Statements

Conditional statements are fundamental to programming as they allow your code to make decisions and execute different blocks of code based on whether a certain condition is `True` or `False`. This control flow is achieved using the `if`, `elif` (short for "else if"), and `else` keywords in Python.

*   **`if` statement:** The most basic conditional statement. The code block following `if` is executed only if its condition evaluates to `True`.

*   **`else` statement:** Used in conjunction with an `if` statement. The code block following `else` is executed if the `if` condition (and any `elif` conditions) evaluates to `False`.

*   **`elif` statement:** Allows you to check multiple conditions sequentially. If the `if` condition is `False`, Python checks the `elif` condition. You can have multiple `elif` blocks. The code block for the first `elif` that evaluates to `True` is executed.

Together, these statements allow you to create powerful logic for handling various scenarios in your programs.

## Control Flow: Loops

Loops are fundamental programming constructs that allow you to execute a block of code repeatedly. This repetition is incredibly useful for automating tasks, processing collections of data, and performing operations until a certain condition is met.

### Why are Loops Important?

*   **Automation:** They allow you to automate repetitive tasks, saving time and reducing the chance of manual errors.
*   **Data Processing:** Essential for working with collections of data (like lists, strings, or numbers), letting you perform operations on each item.
*   **Efficiency:** Instead of writing the same code multiple times, loops enable you to write it once and execute it as many times as needed.

Python provides two primary types of loops: `for` loops and `while` loops.

### `for` Loops

A `for` loop is used for iterating over a sequence (that is, traversing a collection of items, one by one). Sequences can include strings, lists, tuples, dictionaries, and ranges of numbers. The loop continues until it has processed every item in the sequence.

The basic syntax of a `for` loop is:

```python
for item in sequence:
    # code to execute for each item
```

### `while` Loops

A `while` loop is used to repeatedly execute a block of code as long as a given condition remains `True`. The loop continues to run until the condition becomes `False`. It's important to ensure that the condition will eventually become `False` to avoid infinite loops.

The basic syntax of a `while` loop is:

```python
while condition:
    # code to execute as long as the condition is True
    # make sure to include something that changes the condition
```

**Reasoning**:
Now I will add a code cell to demonstrate the 'for' loop with examples for counting, iterating through a string, and summing numbers in a list as instructed.



````
print("--- 'for' Loop Examples ---")

# 5a. Counting from 0 to 4
print("\nCounting from 0 to 4:")
for i in range(5): # range(5) generates numbers 0, 1, 2, 3, 4
    print(i)

# 5b. Iterating through a string
print("\nIterating through a string:")
my_word = "Python"
for char in my_word:
    print(char)

# 5c. Summing numbers in a list
print("\nSumming numbers in a list:")
numbers = [1, 2, 3, 4, 5]
current_sum = 0
for num in numbers:
    current_sum += num # This is equivalent to current_sum = current_sum + num
print(f"The sum of {numbers} is: {current_sum}")
````

**Reasoning**:
Now I will add another code cell to demonstrate the 'while' loop for counting and summing numbers based on a condition, as instructed.



````
print("--- 'while' Loop Examples ---")

# 6a. Counting from 0 to 5
print("\nCounting from 0 to 5:")
count = 0
while count <= 5:
    print(count)
    count += 1 # Important: increment the counter to avoid an infinite loop

# 6b. Summing numbers until a certain condition is met
print("\nSumming numbers until sum exceeds 10:")
current_number = 1
current_sum_while = 0
while current_sum_while <= 10:
    current_sum_while += current_number
    print(f"Adding {current_number}, current sum: {current_sum_while}")
    current_number += 1
````

### Python Data Structures: Lists and Tuples

In Python, **lists** and **tuples** are both ordered collections of items. This means that the elements maintain a specific order in which they were defined, and you can access them by their position (index).

However, their main distinguishing factor is **mutability**:

*   **Lists (`list`)**:
    *   **Mutable**: This means you can change, add, or remove elements from a list after it has been created.
    *   Created using square brackets `[]`.
    *   Often used for collections of items that might need to change during the program's execution.

*   **Tuples (`tuple`)**:
    *   **Immutable**: This means once a tuple is created, you cannot change its elements, add new ones, or remove existing ones.
    *   Created using parentheses `()`.
    *   Often used for collections of items that should not change, such as coordinates, database records, or as keys in dictionaries (which lists cannot be due to their mutability).

````
print("--- Lists and Tuples Demonstration ---")

# 2. Demonstrate how to create lists and tuples
my_list = [10, 20, 30, 40, 50]
my_tuple = (1, 2, 3, 'Python', 5.0)

print(f"\nCreated List: {my_list}")
print(f"Created Tuple: {my_tuple}")

# 3. Access elements of both lists and tuples using indexing
print(f"\nAccessing elements:")
print(f"First element of list (my_list[0]): {my_list[0]}")
print(f"Last element of tuple (my_tuple[-1]): {my_tuple[-1]}")

# 4. Modify elements in a list and append/insert elements
print(f"\nModifying List elements:")
my_list[1] = 25 # Modify an element
print(f"List after modifying my_list[1] to 25: {my_list}")

my_list.append(60) # Append an element
print(f"List after appending 60: {my_list}")

my_list.insert(0, 5) # Insert an element at index 0
print(f"List after inserting 5 at index 0: {my_list}")

# Attempt to modify a tuple element to show its immutability (this will result in an error)
print(f"\nAttempting to modify a Tuple element (will show error):")
try:
    my_tuple[0] = 99
except TypeError as e:
    print(f"Error: {e}")
    print("Tuples are immutable, so elements cannot be changed after creation.")

# 5. Remove elements from a list using remove() and pop()
print(f"\nRemoving List elements:")
my_list.remove(30) # Remove by value
print(f"List after removing value 30: {my_list}")

popped_item = my_list.pop() # Remove and return the last element
print(f"List after popping last item ({popped_item}): {my_list}")

# 6. Iterate through both a list and a tuple using a for loop
print(f"\nIterating through List:")
for item in my_list:
    print(item)

print(f"\nIterating through Tuple:")
for item in my_tuple:
    print(item)

# 7. Provide examples of common list and tuple methods
print(f"\nCommon List Methods:")
print(f"Length of list: {len(my_list)}")
my_list.sort() # Sort the list in ascending order
print(f"List after sorting: {my_list}")
print(f"Count of 25 in list: {my_list.count(25)}")

print(f"\nCommon Tuple Methods:")
print(f"Length of tuple: {len(my_tuple)}")
print(f"Count of 2 in tuple: {my_tuple.count(2)}")
print(f"Index of 'Python' in tuple: {my_tuple.index('Python')}")
````

## Python Data Structures: Dictionaries and Sets

### Subtask:
Introduce dictionaries (unordered, mutable key-value pairs) and sets (unordered, mutable collections of unique elements). Demonstrate creation, access, modification, and common operations for both.


### Dictionaries

A **dictionary** (`dict`) in Python is an unordered collection of data values, used to store data values like a map. Unlike other data types that hold only a single value as an element, Dictionary holds `key:value` pairs.

Key points about dictionaries:

*   **Unordered**: Items in a dictionary do not have a defined order. You cannot access elements by an index (like lists or tuples).
*   **Mutable**: You can add, remove, and modify key-value pairs after the dictionary has been created.
*   **Key-Value Pairs**: Each item in a dictionary consists of a key and its associated value. Keys must be unique and immutable (e.g., strings, numbers, tuples). Values can be of any data type.
*   **Created using curly braces `{}`**: Dictionaries are defined by enclosing a comma-separated list of key-value pairs in curly braces, with a colon `:` separating each key from its value.

**Creating a Dictionary:**

```python
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
```

**Accessing Values:**

You access values in a dictionary using their associated keys, not by numerical index.

```python
print(my_dict["name"])  # Output: Alice
print(my_dict.get("age")) # Output: 30 (Using .get() is safer as it returns None if the key doesn't exist, instead of an error)
```

**Adding and Modifying Key-Value Pairs:**

You can add a new key-value pair or modify an existing one by assigning a value to a key.

```python
my_dict["occupation"] = "Engineer" # Add a new key-value pair
my_dict["age"] = 31               # Modify an existing value
```

**Common Dictionary Methods:**

*   `keys()`: Returns a view object that displays a list of all the keys in the dictionary.
*   `values()`: Returns a view object that displays a list of all the values in the dictionary.
*   `items()`: Returns a view object that displays a list of a dictionary's key-value tuple pairs.
*   `pop(key)`: Removes the item with the specified key and returns its value.
*   `clear()`: Removes all items from the dictionary.


### Sets

A **set** (`set`) in Python is an unordered collection of unique and immutable elements. Sets are highly useful for operations involving uniqueness (e.g., removing duplicates) and mathematical set operations like union, intersection, and difference.

Key points about sets:

*   **Unordered**: Elements in a set do not have a defined order. Like dictionaries, you cannot access elements by an index.
*   **Mutable**: You can add and remove elements from a set after it has been created.
*   **Unique Elements**: Sets automatically handle duplicate elements; if you try to add an element that already exists, it will not be added again.
*   **Immutable Elements**: The elements within a set must be immutable data types (e.g., numbers, strings, tuples). Lists and dictionaries cannot be elements of a set.
*   **Created using curly braces `{}` (or `set()`)**: Sets are defined by enclosing a comma-separated list of elements in curly braces. An empty set must be created using `set()` because `{}` creates an empty dictionary.

**Creating a Set:**

```python
my_set = {1, 2, 3, 4, 4, 5} # Duplicates are automatically removed
empty_set = set() # Correct way to create an empty set
```

**Adding and Removing Elements:**

```python
my_set.add(6)       # Add a single element
my_set.update([7, 8]) # Add multiple elements from an iterable

my_set.remove(3)    # Remove a specific element (raises KeyError if not found)
my_set.discard(10)  # Remove a specific element (does not raise error if not found)
my_set.pop()        # Remove and return an arbitrary element
```

**Common Set Operations:**

Given two sets, `set_a` and `set_b`:

*   **Union (`|` or `union()`)**: Returns a new set containing all unique elements from both sets.
    ```python
    set_a | set_b
    set_a.union(set_b)
    ```
*   **Intersection (`&` or `intersection()`)**: Returns a new set containing only the elements common to both sets.
    ```python
    set_a & set_b
    set_a.intersection(set_b)
    ```
*   **Difference (`-` or `difference()`)**: Returns a new set containing elements present in the first set but not in the second.
    ```python
    set_a - set_b
    set_a.difference(set_b)
    ```
*   **Symmetric Difference (`^` or `symmetric_difference()`)**: Returns a new set containing elements that are in either set, but not in both.
    ```python
    set_a ^ set_b
    set_a.symmetric_difference(set_b)
    ```


**Reasoning**:
Now I will add a code cell to demonstrate all the concepts for both dictionaries and sets as instructed, including creation, access, modification, common methods, and iteration.



````
print("--- Dictionaries and Sets Demonstration ---")

# 1. Create example dictionaries and sets
my_dict = {
    "name": "Bob",
    "age": 25,
    "city": "Chicago"
}
my_set = {10, 20, 30, 20, 40} # Duplicates (20) will be automatically removed

print(f"\nCreated Dictionary: {my_dict}")
print(f"Created Set: {my_set}")

# 2. Accessing values from a dictionary using keys
print(f"\nAccessing Dictionary values:")
print(f"Name: {my_dict['name']}")
print(f"Age (using .get()): {my_dict.get('age')}")

# 3. Adding new key-value pairs and modifying existing ones in a dictionary
print(f"\nModifying Dictionary:")
my_dict["email"] = "bob@example.com" # Add new key-value pair
my_dict["age"] = 26 # Modify existing value
print(f"Dictionary after modification: {my_dict}")

# 4. Adding and removing elements from a set
print(f"\nModifying Set:")
my_set.add(50) # Add a new element
my_set.remove(10) # Remove an existing element
print(f"Set after modification: {my_set}")

# 5. Demonstrating common dictionary methods
print(f"\nCommon Dictionary Methods:")
print(f"Dictionary Keys: {my_dict.keys()}")
print(f"Dictionary Values: {my_dict.values()}")
print(f"Dictionary Items: {my_dict.items()}")
popped_value = my_dict.pop("city")
print(f"Dictionary after popping 'city': {my_dict}, Popped Value: {popped_value}")

# 6. Demonstrating common set operations
print(f"\nCommon Set Operations:")
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
print(f"Set A: {set_a}, Set B: {set_b}")
print(f"Union (set_a | set_b): {set_a | set_b}")
print(f"Intersection (set_a & set_b): {set_a & set_b}")
print(f"Difference (set_a - set_b): {set_a - set_b}")
print(f"Symmetric Difference (set_a ^ set_b): {set_a ^ set_b}")

# 7. Iterating through a dictionary (keys, values, or items) and a set
print(f"\nIterating through Dictionary Keys:")
for key in my_dict:
    print(key)

print(f"\nIterating through Dictionary Items:")
for key, value in my_dict.items():
    print(f"{key}: {value}")

print(f"\nIterating through Set:")
for element in my_set:
    print(element)
````

## Functions: Organizing Your Code

### Functions: Organizing Your Code

**Functions** are self-contained blocks of code that perform a specific task. They are fundamental for writing clean, efficient, and maintainable Python programs.

#### Importance of Functions:

1.  **Code Reusability**: Once a function is defined, it can be called multiple times throughout your program (or even in different programs), avoiding redundant code. This adheres to the "Don't Repeat Yourself" (DRY) principle.
2.  **Organization and Readability**: Functions break down complex problems into smaller, manageable pieces, making your code easier to understand, debug, and maintain.
3.  **Modularity**: Functions promote modular programming, where different parts of your code are independent and can be developed and tested separately.
4.  **Abstraction**: Functions allow you to hide the complex details of an operation behind a simple name, letting you focus on *what* the function does rather than *how* it does it.

### Error Handling: `try-except` Blocks

In programming, **errors** are inevitable. They can be broadly categorized into:
*   **Syntax Errors:** These occur when the rules of the language are violated (e.g., typos, missing colons).
*   **Runtime Errors (Exceptions):** These occur when the program is syntactically correct but encounters a problem during execution (e.g., trying to divide by zero, accessing a file that doesn't exist, converting invalid data).

**Error handling** is the process of anticipating, detecting, and resolving these runtime errors to prevent your program from crashing abruptly. Python provides a powerful mechanism for this using `try` and `except` blocks.

#### The Importance of `try-except` Blocks:

*   **Graceful Degradation:** Instead of crashing, your program can respond to errors in a controlled manner, providing useful feedback to the user or logging the error for debugging.
*   **Robustness:** Makes your applications more reliable and resilient to unexpected input or circumstances.
*   **User Experience:** Prevents cryptic error messages from being shown to end-users.

#### Basic Syntax of `try-except`:

The `try` block contains the code that might raise an exception. If an exception occurs within the `try` block, Python immediately stops executing the rest of the `try` block and looks for a matching `except` block.

The `except` block specifies how to handle a particular type of exception. If the type of exception raised in the `try` block matches the exception type specified in the `except` block, the code within that `except` block is executed.

```python
try:
    # Code that might raise an exception
    # For example, a division operation or type conversion
    result = 10 / 0 # This will raise a ZeroDivisionError
except ZeroDivisionError:
    # Code to execute if a ZeroDivisionError occurs
    print("Error: Cannot divide by zero!")
except ValueError:
    # Code to execute if a ValueError occurs
    print("Error: Invalid value provided.")
except Exception as e: # Generic exception handler (catches any other error)
    print(f"An unexpected error occurred: {e}")
else:
    # (Optional) Code to execute if the try block completes WITHOUT an exception
    print("Operation successful!")
finally:
    # (Optional) Code that ALWAYS executes, regardless of whether an exception occurred or not
    print("Execution of try-except block finished.")
```

### Modules and Packages

As your Python programs grow in size and complexity, it becomes essential to organize your code to keep it manageable and reusable. This is where **modules** and **packages** come in handy.

#### What are Modules?

A **module** is simply a file containing Python code (variables, functions, classes, etc.). The file name is the module name with the `.py` extension. Modules allow you to logically organize your Python code, group related code into a file, and make your code reusable.

For example, if you have a file named `my_functions.py` containing a function `greet()`, then `my_functions` is a module.

#### What are Packages?

A **package** is a way of organizing related modules into a directory hierarchy. Essentially, a package is a directory containing multiple module files and a special `__init__.py` file (which can be empty). This structure allows for grouping modules into a namespace, preventing name collisions, and making it easier to manage larger projects.

Think of a module as a single book, and a package as a library containing many related books (modules).

#### Why Use Modules and Packages?

*   **Code Organization:** Breaks down large programs into smaller, manageable, and logical units.
*   **Reusability:** Code written in one module can be easily used in other modules or programs.
*   **Readability:** Makes code easier to understand and maintain.
*   **Namespace Management:** Helps avoid naming conflicts by providing separate namespaces for each module.

#### How to Import and Use Them (`import` statement)

The `import` statement is used to bring modules or specific parts of modules into your current Python script or environment. There are several ways to use it:

1.  **Import an entire module:**
    This imports the whole module, and you access its contents using `module_name.item_name`.
    ```python
    import math
    print(math.pi)
    ```

2.  **Import specific functions or variables from a module:**
    This allows you to import only the necessary parts, and you can use them directly without the `module_name.` prefix.
    ```python
    from math import pi, sqrt
    print(pi)
    print(sqrt(16))
    ```

3.  **Import all contents from a module:**
    Using `*` imports all public names from the module directly into the current namespace. This is generally discouraged as it can lead to name clashes if multiple modules have items with the same name.
    ```python
    from math import *
    print(sin(0))
    ```

4.  **Import a module with an alias:**
    You can give an imported module a shorter or more convenient name using the `as` keyword. This is very common for popular libraries like `numpy` (often imported as `np`) and `pandas` (as `pd`).
    ```python
    import random as rnd
    print(rnd.randint(1, 10))
    ```

Understanding how to effectively use modules and packages is crucial for writing robust and organized Python applications.
